```python
import xarray as xr
```

```python
ds = xr.open_dataset('data.nc').sel(expver=1)
```

```python
#convert to pandas data frame
df = ds.to_dataframe()
df
```

```python
da = ds.time.dt.month
da.to_dataframe()
```

```python
da.to_series()
```

```python
xr.Dataset.from_dataframe(df)
```

```python
xr.DataArray.from_series(df.skt)
```

```python
ds.to_netcdf('export.nc')
```

```python
xr.open_dataset('export.nc')
```
